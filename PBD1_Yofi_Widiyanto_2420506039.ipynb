# 2. Pengenalan dan Praktik Dasar PySpark dan Pandas
- **Tugas 1**: Jalankan kode di atas dan buat modifikasi dengan menambahkan data lain berupa kolom pekerjaan, hobi dan gender:

```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana menambahkan kolom pekerjaan, hobi, dan gender
data = [("Ali", 34, "Dokter", "Membaca", "L"),
        ("Budi", 23, "Guru", "Olahraga", "L"),
        ("Citra", 29, "Insinyur", "Memasak", "P"),
        ("Dina", 45, "Perawat", "Menulis", "P")]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df_spark = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
print("DataFrame Pyspark:")
df_spark.show()
```

# 3. Praktik PySpark Lanjutan
- **Tugas 2**: Lakukan filter, penghitungan rata-rata, dan pengurutan data menggunakan PySpark:

```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34, "Dokter", "Membaca", "L"),
        ("Budi", 23, "Guru", "Olahraga", "L"),
        ("Citra", 29, "Insinyur", "Memasak", "P"),
        ("Dina", 45, "Perawat", "Menulis", "P")]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df_spark = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
print("DataFrame Pyspark:")
df_spark.show()

# Melakukan filter data usia di atas 30
print("DataFrame hasil filtering usia di atas 30:")
df_spark_filtered = df_spark.filter(df_spark['Usia'] > 30)
df_spark_filtered.show()

# Melakukan penghitungan rata-rata usia
from pyspark.sql.functions import avg
print("Rata-rata usia:")
df_spark.groupBy().agg(avg("Usia")).show()

# Melakukan pengurutan data berdasarkan usia
print("DataFrame diurutkan berdasarkan usia:")
df_spark_sorted= df_spark.orderBy("Usia", ascending=False)
df_spark_sorted.show()
```

# 4. Praktik dengan Pandas
- **Tugas 3**: Modifikasi DataFrame Pandas dengan menambahkan kolom baru dan melakukan operasi seperti filtering data berdasarkan usia:

```python
import pandas as pd

# Membuat DataFrame Pandas
data_pandas = {"Nama": ["Mutiara", "Sovia", "Soim", "Haru"],
               "Usia": [20, 21, 19, 25],
               "Pekerjaan": ["Pengusaha", "Guru", "Pedagang", "Petualang"],
               "Hobi": ["Membaca", "Membaca", "Jualan", "Tinju"],
               "Gender": ["P", "P", "P", "P"]}
df_pandas = pd.DataFrame(data_pandas)
# Menampilkan DataFrame Pandas
print("DataFrame Pandas:")
display(df_pandas)

#Melakukan filtering data berdasarkan usia di bawah 20
df_pandas_filtered = df_pandas[df_pandas["Usia"] < 20].reset_index(drop=True)
print("\nDataFrame Pandas usia di bawah 20:")
display(df_pandas_filtered)
```

# 5. Praktik Pandas Lanjutan
- **Tugas 4**: Lakukan penggabungan DataFrame dan visualisasikan data dengan Pandas.

```python
import pandas as pd

# Membuat DataFrame Pandas
data_pandas = {"Nama": ["Mutiara", "Sovia", "Soim", "Haru"],
               "Usia": [20, 21, 19, 25]}
df_pandas_1 = pd.DataFrame(data_pandas)

# Membuat DataFrame kedua
data_pandas_2 = {"Nama": ["Mutiara", "Sovia", "Soim", "Haru"],
                 "Pekerjaan": ["Pengusaha", "Guru", "Pedagang", "Petualang"]}
df_pandas_2 = pd.DataFrame(data_pandas_2)

# Join antara dua DataFrame
df_joined = pd.merge(df_pandas_1, df_pandas_2, on="Nama")
print("Data Pandas:")
display(df_joined)

# Menghitung statistik deskriptif
print("\nStastistik Deskriptif:")
display(df_pandas.describe())

# Plotting Data
import matplotlib.pyplot as plt
df_pandas['Usia'].plot(kind='bar')
print("\nData Visual dengan Diagram:")
plt.show()
```

# 5. Menggabungkan PySpark dan Pandas
- **Tugas 5**: Gunakan metode ini untuk menggabungkan data yang Anda buat di PySpark dengan data dari Pandas, kemudian lakukan analisis sederhana seperti menghitung rata-rata usia.

```python
# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df_spark.toPandas()

# Mengonversi DataFrame dari Pandas ke PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Menampilkan DataFrame hasil konversi
print("Hasil dari DataFrame Pyspark ke Pandas:")
display(df_pandas_from_spark)
print("\nHasil dari DataFrame Panads ke Pyspark:")
df_spark_from_pandas.show()

# Data gabungan Pyspark dan Pandas
combined = pd.concat([df_pandas_from_spark, df_pandas]).reset_index(drop=True)
print("\nDataFrame Gabungan:")
display(combined)

# Nilai Rata-rata ke dua DataFrame Pandas
print(f"\nPerhitungan usia rata-rata:", combined["Usia"].mean())
```

# 6. Konversi Data antara PySpark dan Pandas
- **Tugas 6**: Gabungkan data dari PySpark dan Pandas, lalu lakukan operasi statistik seperti menghitung nilai maksimum usia.

```python
# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df_spark.toPandas()

# Mengonversi DataFrame dari Pandas ke PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Menampilkan DataFrame hasil konversi
print("Hasil dari DataFrame Pyspark ke Pandas:")
display(df_pandas_from_spark)
print("\nHasil dari DataFrame Panads ke Pyspark:")
df_spark_from_pandas.show()

# Data gabungan Pyspark dan Pandas
combined = pd.concat([df_pandas_from_spark, df_pandas]).reset_index(drop=True)
print("\nDataFrame Gabungan:")
display(combined)

# Menghitung nilai Max usia
print(f"\nNilai maksimum usia adalah: ", combined["Usia"].max())
```
