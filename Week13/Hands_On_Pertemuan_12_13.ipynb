## Pertemuan 12: Data Cleaning and Preparation using Pandas

### Topics Covered

- Data transformation and normalization.
- Data filtering and deduplication.
- Standardization of categorical data.
- Outlier detection and handling.


```python
# Exercise 1: Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Data
months = [
    "Jan","Feb","Mar","Apr","May","Jun",
    "Jul","Aug","Sep","Oct","Nov","Dec"
]

gow = [12000,14000,15000,15500,16000,15800,16200,17000,17500,18000,18500,20000]
gta = [8000,8500,9000,9200,9500,9700,10000,10200,10500,11000,11500,12000]
monhunt = [15000,16000,17000,16500,18000,17500,19000,19500,20000,21000,21500,23000]

# posisi bar
x = np.arange(len(months))
width = 0.25

plt.figure(figsize=(14, 7))

plt.bar(x - width, gow, width, label="God of War")
plt.bar(x, gta, width, label="Grand Thieft Auto")
plt.bar(x + width, monhunt, width, label="Monster Hunter")

plt.xticks(x, months)
plt.ylabel("Sales")
plt.title("Monthly Game Sales")
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()

df = pd.DataFrame({
    "Month": months,
    "God_of_War": gow,
    "Grand_Thieft_Auto": gta,
    "Monster_Hunter": monhunt
})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Month", y="Monster_Hunter", s=120, color="brown")

plt.title("Monster Hunter Monthly Sales")
plt.grid(alpha=0.3)
plt.show()

sns.set_style("whitegrid")
sns.set_palette("Set2")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1 ‚Äì line chart
axes[0].plot(months, gow, marker="o")
axes[0].set_title("God of War Trend")

# Plot 2 ‚Äì scatter plot
sns.scatterplot(ax=axes[1], data=df, x="Month", y="Grand_Thieft_Auto", s=120)
axes[1].set_title("GTA Sales")

plt.tight_layout()
plt.show()
```

```python
# Task Pertemuan 12
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("diabetes.csv")
df.head()

print("\nMissing Values before Cleaning:")
print(df.isnull().sum())

cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in cols_with_zero:
    df[col] = df[col].replace(0, np.nan)

print("\nMissing values 0 ‚Üí NaN:")
print(df.isnull().sum())

df[cols_with_zero] = df[cols_with_zero].fillna(df[cols_with_zero].median())

def detect_outliers_iqr(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] < lower) | (df[col] > upper)]

outliers_bmi = detect_outliers_iqr("BMI")
print("\nJumlah outlier pada BMI:", len(outliers_bmi))

for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
              np.where(df[col] > upper, upper, df[col]))

num_cols = df.columns.drop("Outcome")

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print("\nData setelah normalisasi:")
display(df.head())

plt.figure(figsize=(9,3))
df[num_cols].hist(bins=20, figsize=(12,8))
plt.suptitle("Distribusi Data Setelah Cleaning & Normalisasi")
plt.show()
```

```python
# Pertemuan 13
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('diabetes.csv')

cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in cols_with_zero:
    df[col] = df[col].replace(0, np.nan)

df[cols_with_zero] = df[cols_with_zero].fillna(df[cols_with_zero].median())

def detect_outliers_iqr(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] < lower) | (df[col] > upper)]


for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
              np.where(df[col] > upper, upper, df[col]))

num_cols = df.columns.drop("Outcome")

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

sns.set_style("whitegrid")
sns.set_palette("Set2")

plt.figure(figsize=(18, 12))

plt.subplot(2, 2, 1)
sns.countplot(data=df, x="Outcome")
plt.title("Outcome Distribution")
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")

plt.subplot(2, 2, 2)
sns.histplot(df["Age"], kde=True, bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")

plt.subplot(2, 2, 3)
sns.scatterplot(data=df, x="Glucose", y="BMI", hue="Outcome", s=80)
plt.title("Glucose vs BMI")
plt.xlabel("Glucose Level")
plt.ylabel("BMI")

plt.subplot(2, 2, 4)
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=False, cmap="viridis")
plt.title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```
